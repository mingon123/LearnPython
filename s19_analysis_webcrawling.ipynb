```python
#웹 크롤링이란 웹의 정보를 자동으로 수집하는 것을 의미하며 이런 목적을 위해 만든 프로그램을 웹 크롤러라고 말함
#http://quotes.toscrape.com/ 에서 명언을 수집하는 웹 크롤러 작성
#뷰티풀수프 설치하기
#뷰티풀수프는 웹 문서를 구성하는 HTML과 XML 문서에서 원한느 정보를 쉽게 추출할 수 있는 모듈을 모아놓음
#파이썬 패키지
#(python_env)C:\Class\python_work>conda install beautifulsoup4  (터미널->cmd에서 입력,base로 되어있으면 conda activate python_env로 활성화)
#웹에서 얻은 데이터를 다루는 파이썬 패키지
import urllib.request as ur
from bs4 import BeautifulSoup as bs

#웹 문서 자료를 가져와 가공하기 (아래는 명언 사이트)
url = 'http://quotes.toscrape.com/'
#urlopen으로 웹사이트 정보 가져오기
html = ur.urlopen(url)
print(type(html))
```

```python
print(html)
```

```python
#b''는 파이썬에서 웹으로부터 받은 데이터가 아직 문자 인코딩이 적용되지 않은 바이트 형태의 원시 데이터
byte_content = html.read()
print(byte_content)
print('-' * 50)

#UTF-8으로 디코딩
decoded_content = byte_content.decode('utf-8')
print(decoded_content)
```

```python
#정보를 100개까지만 슬라이싱을 해서 정보 읽기
print(decoded_content[:100])
```

```python
#뷰티풀수프로 자료형 변환하기(html 문자열에 대해서,html파싱)
soup = bs(ur.urlopen(url).read(),'html.parser')
print(type(soup))
```

```python
print(soup)
```

```python
#find_all로 원하는 태그만 모아서 리스트 반환
print(soup.find_all('span'))
```

```python
quote = soup.find_all('span')
print(quote[0]) #리스트의 첫 번째 요소의 값만 반환
print(quote[0].text) #텍스만 출력
```

```python
#모든 요소에서 텍스트만 추출
for i in quote:
    print(i.text)
```

```python
#<div>태그 안에 정의된 특정 클래스 찾기, 리스트로 반환
print(soup.find_all('div',{"class":"quote"}))
```

```python
#requests 라이브러리를 통해 웹 요청을 보내고, BeautifulSouop를 통해 HTML를 파싱하기 위한 라이브러리 임포트
import requests
from bs4 import BeautifulSoup

#검색 URL (박스오피스)
url = "https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=1&ie=utf8&query=%EB%B0%95%EC%8A%A4%EC%98%A4%ED%94%BC%EC%8A%A4"

#요청 및 HTML 파싱
response = requests.get(url)
print(response)
print('-' * 50)

#응답 받은 HTML을 BeautifulSoup 객체로 파싱
soup = BeautifulSoup(response.text,"html.parser")
print(soup)
```

```python
#영화 리스트 선택
#CSS선택자 기반으로 박스오피스 영화 항목들을 선택
#'div.mflick' 내부에 있는 각 'div.item' 요소들이 영화 하나에 해당
items = soup.select("div.mflick div.item")
print(items)
```

```python
#결과 저장 리스트
cinema_list = []

#최대 30개까지 크롤링
for content in items[:30]:
    print(content)

    #제목 정보가 들어있는 태그 선택(div.title_box 내부의 .name클래스)
    title_tag = content.select_one("div.title_box .name")
    #영화 이미지가 들어있는 img 태그 선택
    image_tag = content.select_one("img")
    #관객 수 등 부가정보가 들어있는 태그 선택(div.title_box 내부의 .sub_text 클래스)
    audience_tag = content.select_one("div.title_box .sub_text")

    #세 개의 정보가 모두 존재할 경우만 처리
    if title_tag and image_tag and audience_tag:
        cinema = {
            "title":title_tag.get_text(strip=True), #strip=True는 앞뒤 공백만 제거.중간의 공백은 그대로 유지. - 텍스트읽어옴
            "audience":audience_tag.get_text(strip=True), #텍스트읽어옴
            "image_url":image_tag.get("src") #속성읽어옴
        }
        #딕셔너리 형태로 저장한 영화 정보를 리스트에 추가
        cinema_list.append(cinema)
```

```python
#최종 수집된 영화 정보 출력
for movie in cinema_list:
    print(movie)
```
